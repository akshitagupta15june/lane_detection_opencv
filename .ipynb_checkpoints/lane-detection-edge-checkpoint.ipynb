# Selecting region of interest

```python
def region_of_interest(image):
    height = image.shape[0]
    polygons= np.array([
                            [(44, height), (129, height), (115, 20)]
                        ])
    mask = np.zeros_like(image)
    cv2.fillPoly(mask, polygons,255)
    segment = cv2.bitwise_and(image, mask)
    return segment
plt.imshow(region_of_interest(canny),cmap='gray')
```

# using hough transform 

```python
lines = cv2.HoughLinesP(region_of_interest(canny), 2, np.pi / 180, 25, np.array([]), minLineLength = 40, maxLineGap = 40)
print(lines)
```

```python
#function to display lines detected by hough transform

def display_lines(image,lines):
    line_image = np.zeros_like(image)
    for line in lines:
        x1,y1,x2,y2 = line.reshape(4)
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),4)
    return line_image
```

```python
laneimg = np.copy(img)
gray = cv2.cvtColor(laneimg, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray,(5,5),0)
canny = cv2.Canny(blur,50,150)
#plt.imshow(canny,cmap='gray')
cropped_img = region_of_interest(canny)
lines = cv2.HoughLinesP(cropped_img, 2, np.pi / 180, 25, np.array([]), minLineLength = 40, maxLineGap = 40)
print(lines)
line_image = display_lines(laneimg,lines)
overlay_img = cv2.addWeighted(laneimg,0.8,line_image,1,1)

plt.figure(2,figsize=(10,10))
plt.subplot(211)
plt.imshow(cropped_img,cmap='gray')

plt.subplot(212)
plt.imshow(overlay_img)
plt.show()
```

# Separating lanes

```python
right_lane=[]
left_lane=[]
for i in lines:
    #print(i)
    if (i[0][0]>100 and i[0][2]>100 ):
        right_lane.append(i[0])
    else:
        left_lane.append(i[0])   
right_lane=np.array(right_lane)
left_lane=np.array(left_lane)
print(right_lane,"\n\n")
print(left_lane)

right_line_image = display_lines(laneimg,right_lane)
left_line_image = display_lines(laneimg,left_lane)

plt.figure(2,figsize=(10,10))
plt.subplot(211)
plt.imshow(right_line_image)

plt.subplot(212)
plt.imshow(left_line_image)
plt.show()
plt.imshow(right_line_image+left_line_image)
```

# Finding best fitting lane lines

```python
x_right=[]
y_right=[]
for i in right_lane:
    x_right.append(i[0])
    x_right.append(i[2])
    
    y_right.append(i[1])
    y_right.append(i[3])
x=len(x_right)
for j in range(x - 1, -1, -1):
    if(x_right[j]<119):
        x_right.remove(x_right[j])
        y_right.remove(y_right[j])
    x=len(x_right)
        
print(len(x_right))
print("x_right  ",x_right)
print("y_right  ",y_right)
```

```python
plt.scatter(x_right,y_right)
m1, b1 = np.polyfit(x_right, y_right, 1)
print(m1,b1)
y_right_bestfit=[]
for i in range(len(x_right)):
    y_right_bestfit.append(m1*x_right[i]+b1)
    
print(x_right,y_right_bestfit)
right_bestfit=[[118,int((m1*118.5)+b1),127,int((m1*127)+b1)]]
right_bestfit=np.array(right_bestfit)
print(right_bestfit)
plt.plot(x_right, y_right_bestfit)
```

```python
x_left=[]
y_left=[]
for i in left_lane:
    x_left.append(i[0])
    x_left.append(i[2])
    
    y_left.append(i[1])
    y_left.append(i[3])
print("x_left  ",x_left)
print("y_left  ",y_left)
```

```python
plt.scatter(x_left,y_left)
m2, b2 = np.polyfit(x_left, y_left, 1)
y_left_bestfit=[]
for i in range(len(x_left)):
    y_left_bestfit.append(m2*x_left[i]+b2)
    
print(x_left,y_left_bestfit)
left_bestfit=[[x_left[0],int(y_left_bestfit[0]),106,int((m2*106)+b2)]]
left_bestfit=np.array(left_bestfit)
print(left_bestfit)
plt.plot(x_left, y_left_bestfit)
```

# Displaying best fitting lane lines

```python
right = display_lines(laneimg,right_bestfit)
left = display_lines(laneimg,left_bestfit)

plt.figure(2,figsize=(10,10))
plt.subplot(211)
plt.imshow(right)

plt.subplot(212)
plt.imshow(left)
plt.show()
plt.imshow(right+left)
```

```python
overlay_img_new = cv2.addWeighted(laneimg,0.8,right+left,1,1)
plt.imshow(overlay_img_new)
```

# Filling area b/w lanes

```python
overlay_img_new_copy= np.copy(overlay_img_new)
def fill_region_of_interest(img,poly):
    mask = np.zeros_like(img)
    match_mask_color = (0,0,255)
    cv2.fillPoly(mask,poly, match_mask_color)
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
lane_vertices = [(right_bestfit[0][0],right_bestfit[0][1]),
                        (right_bestfit[0][2],right_bestfit[0][3]),
                        (left_bestfit[0][0],left_bestfit[0][1]),
                        (left_bestfit[0][2],left_bestfit[0][3])]
```

```python
fillarea = fill_region_of_interest(
    overlay_img_new_copy,
    np.array([lane_vertices], np.int32)
)
plt.figure()
plt.imshow(fillarea)
plt.show()
```

```python
final_img = overlay_img_new_copy+fillarea
plt.imshow(final_img);
```
